```python
# sex plot:
sns.factorplot("sex", data=titanic, kind="count")
```

```python
sns.factorplot("class", data=titanic, kind="count")
```

```python
# plot for passenger classes according to sex:
sns.factorplot("class", data=titanic, hue="sex", kind="count")
plt.title("Class by Sex")
sns.set_style("dark")
```

```python
#passenger survival according to sex:
sns.factorplot("sex", data=titanic, hue="alive", kind="count")
plt.title("Survival by Sex")
sns.set_style("dark")
```

```python
# removing deck column:
# inplace means that deletion is permanent
titanic.drop("deck", axis=1, inplace=True)
```

```python
titanic.info()
```

```python
sns.factorplot("embarked", data=titanic, kind="count")
```

```python
#mean of the ages
titanic["age"] = titanic["age"].fillna(titanic["age"].mean())
```

```python
sns.factorplot("embark_town", data = titanic, hue = "class", kind = "count")
```

```python
sns.factorplot("survived", data = titanic, hue = "class", kind = "count")
```

# Different Types of Plots:
- distribution plots
 - a.Dist Plot
 - b. Joint Plot
 - c. Pair Plot
 - d. Rug Plot
- Categorical Plots
 - a. Bar Plot
 - b. Count Plot
 - c. Box Plot
 - d. Violin Plot

```python
sns.distplot(x=titanic["age"], bins=10)
```

```python
# joint plot
# combination of distplot of two variables
# bivariate analysis

#method1 = scatter
sns.jointplot(x=titanic["class"], y=titanic["fare"], kind="scatter")
```

```python
# joint plot - 2
sns.jointplot(x=titanic["age"], y=titanic["fare"], kind="hex")
```

```python
# c pair-plot
# all numerical attributes of the data
# and plot the scatterplot for two different variables

sns.pairplot(titanic)
```

```python
# rugplot
# univariate analysis:
sns.rugplot(x=titanic["fare"])
```

### Categorical plots

```python
# barplot is for bivariate analysis
# x-axis = categorical
# y-axis = continuous
sns.barplot(x=titanic["pclass"],y=titanic["fare"])
```

```python
#b. countplot:
# univariate analysis
# number of occurences (categorical variables)
sns.countplot(titanic["pclass"])
```

```python
#c. boxplot
# continuous variables
# 5 points are shown: max, min, mean, 1st quartile, 3rd quartile
# info about outliers
sns.boxplot(titanic["pclass"])
```

```python
sns.boxplot(x=titanic["sex"], y=titanic["age"])
```

```python
# d. violin plot
sns.violinplot(x=titanic["sex"], y=titanic["age"])
```

---
- Advanced Plots

```python
# strip Plot
# continuous and categorical
sns.stripplot(y=titanic["age"], x=titanic["pclass"])
```

```python
# Swarm Plot
# combo of strip plot and violin plot
sns.swarmplot(y=titanic["age"], x=titanic["pclass"])
```
